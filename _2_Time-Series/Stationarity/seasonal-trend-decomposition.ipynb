```python
import pandas as pd

from IPython.display import display
from statsmodels.tsa.seasonal import DecomposeResult, seasonal_decompose
```

## A non-stationary time-series

```python
df: pd.DataFrame = pd.read_csv('../Resources/airline-passengers.csv')
display(df.head())
```

```python
df['Month'] = df['Month'].apply(lambda x: pd.to_datetime(x))
df = df.set_index('Month')
```

```python
result: DecomposeResult = seasonal_decompose(df, model='multiplicative')
result.plot()
```

## A stationary Time-Series

```python
df: pd.DataFrame = pd.read_csv('../Resources/daily-total-female-births.csv')
display(df.head())
```

```python
df['Date'] = df['Date'].apply(lambda x: pd.to_datetime(x))
df = df.set_index('Date')
```

```python
result: DecomposeResult = seasonal_decompose(df, model='multiplicative')
result.plot()
```
